```python
import requests

result = requests.get('http://cbr.ru')
html = result.text
print(html)
```

```python
import re

match = re.search(r'Евро\D+(\d+,\d+)', html)
rate = match.group(1)
print(rate)
```

```python
html = 'Курс евро на сегодня 68,7614, курс евро на завтра 67,8901'
match = re.search(r'Евро\D+(\d+,\d+)', html, re.IGNORECASE)
rate = match.group(1)
rate
```
